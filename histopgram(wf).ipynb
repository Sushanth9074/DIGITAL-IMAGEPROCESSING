```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img=cv2.imread('/content/butterfly.png')
(row,col)=img.shape[:2]
```

```python
y=np.zeros((256,),dtype=int)
for i in range(0,row):
    for j in range(0,col):
        a=img[i,j]
        y[a+1]=y[a+1]+1
```

```python
plt.plot(y)
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
